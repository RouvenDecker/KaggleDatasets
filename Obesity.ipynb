```python
X = df.drop(columns=["Obesity"])
y = df["Obesity"]
```

##### Datasplit

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

##### Oversampling

```python
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler()
x_train,y_train = ros.fit_resample(x_train,y_train)
```

```python
print(type(y_train))
```

```python
from sklearn.compose import make_column_selector

categorical_features = make_column_selector(dtype_include=["object","category"])
numeric_features = make_column_selector(dtype_include=["float64","int64"])
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import ColumnTransformer

num_pipe = Pipeline([
    ("sscaler",StandardScaler())
])

cat_pipe = Pipeline([
    ("oht",OneHotEncoder(sparse_output=False))
])

ctrf = ColumnTransformer([
    ("num_pipe",num_pipe,numeric_features),
    ("cat_pipe",cat_pipe,categorical_features)
])

ctrf.fit(x_train)

x_train_trf = ctrf.transform(x_train)
x_test_trf = ctrf.transform(x_test)
```

```python
x_train_trf[0]
```

```python
from sklearn.preprocessing import LabelEncoder


le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_test_enc = le.transform(y_test)

print(y_train_enc[:4])
le.classes_
```

#### Dummy Baseline

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score,f1_score

dummy_clf = DummyClassifier(strategy="stratified")
dummy_clf.fit(x_train_trf,y_train_enc)

y_pred = dummy_clf.predict(x_test_trf)
acc = accuracy_score(y_test_enc,y_pred)
f1 = f1_score(y_test_enc,y_pred,average="macro")

print(f"acc: {acc}")
print(f"f1: {f1}")
```

```python
print(100 / 7)
```

#### KNN 

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

params = {
    "n_neighbors":[4,5,7,9],
    "weights":["uniform","distance"]
}

clf = KNeighborsClassifier()

grid = GridSearchCV(
    estimator=clf,
    param_grid=params,
    cv=5,
    n_jobs=4
)
grid.fit(x_train_trf,y_train_enc)
```

```python
best_estimator = grid.best_estimator_
best_params = grid.best_params_
best_score = grid.best_score_

print(f"best estimator: {best_estimator}")
print(f"best_params: {best_params}")
print(f"best score: {best_score}")
```

```python
clf = KNeighborsClassifier(**best_params)
clf.fit(x_train_trf,y_train_enc)
y_pred = clf.predict(x_test_trf)
```

```python
print(le.classes_)
print(y_pred[:5])
print(y_test_enc[:5])
```

```python
from sklearn.metrics import classification_report

report = classification_report(
    y_true=y_test_enc,
    y_pred=y_pred)
print(report)
print(le.classes_)
```

#### Tree


```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()
params={
    "max_depth":[8,10,12,15],
    "min_samples_leaf":[1,3]
}

grid = GridSearchCV(
    estimator=clf,
    param_grid=params,
    cv=5
)

grid.fit(x_train_trf,y_train_enc)
```

```python
best_estimator = grid.best_estimator_
best_params = grid.best_params_
best_score = grid.best_score_

print(f"best estimator: {best_estimator}")
print(f"best_params: {best_params}")
print(f"best score: {best_score}")
```

```python
clf = DecisionTreeClassifier(**best_params)
clf.fit(x_train_trf,y_train_enc)
y_pred = clf.predict(x_test_trf)
```

```python
report = classification_report(
    y_true=y_test_enc,
    y_pred=y_pred)
print(report)
print(le.classes_)
```

#### RandomForest

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier()
params={
    "n_estimators":[200,300,400],
    "max_depth":[15,20],
    "min_samples_leaf":[1,2]
}


grid = GridSearchCV(
    estimator=clf,
    param_grid=params,
    cv=5
)

grid.fit(x_train_trf,y_train_enc)
```

```python
best_estimator = grid.best_estimator_
best_params = grid.best_params_
best_score = grid.best_score_

print(f"best estimator: {best_estimator}")
print(f"best_params: {best_params}")
print(f"best score: {best_score}")
```

```python
clf = RandomForestClassifier(**best_params)
clf.fit(x_train_trf,y_train_enc)
y_pred = clf.predict(x_test_trf)
```

```python
report = classification_report(
    y_true=y_test_enc,
    y_pred=y_pred)
print(report)
print(le.classes_)
```
