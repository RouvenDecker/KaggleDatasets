```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("fivethirtyeight")
import seaborn as sns

wine = pd.read_csv("data/winequality-red.csv")
print(wine.head())
```

```python
wine.info()
```

```python
wine.describe()
```

### Correlation

```python
corr = wine.corr(numeric_only=True)
plt.figure(figsize=(10, 5))
sns.heatmap(data=corr, annot=True)
```

### Feature distribution

```python
fig, ax = plt.subplots(1, 3, sharex="col")
fig.set_size_inches(6, 3)
for i in range(3):
    ax[i].boxplot(
        wine.iloc[:, i],
        label=wine.columns[i],
        patch_artist=True,
        flierprops={"markerfacecolor": "red", "color": "red", "markersize": 8},
    )
    ax[i].set_title(wine.columns[i], color="black", fontsize=10)
```

```python
fig, ax = plt.subplots(1, 3, sharex="col")
fig.set_size_inches(6, 3)
for i in range(3):
    ax[i].hist(wine.iloc[:, i], label=wine.columns[i], bins=30, alpha=0.4)
    ax[i].set_title(wine.columns[i], color="black", fontsize=10)
```
