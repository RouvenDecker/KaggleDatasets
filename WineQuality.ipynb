## Target distribution

```python
plt.figure(figsize=(5, 3))
sns.histplot(wine, x="quality")
```

```python
wine["quality"].value_counts()
```

### combine classes

```python
quality_map = {3: "low", 4: "low", 5: "medium", 6: "medium", 7: "high", 8: "high"}

wine["quality"] = wine["quality"].map(quality_map)
```

```python
wine["quality"].value_counts()
```

```python
y = wine.pop("quality").to_numpy()
X = wine.to_numpy()
print(X[:2])
print(y[:2])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler


pipe = Pipeline([("scaler", StandardScaler())])


X_train_trf = pipe.fit_transform(X_train)
X_test_trf = pipe.transform(X_test)
```

```python
X_train_trf.shape
```

```python
y_train.shape
```

#### oversampling of class "low"

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy={"low": 300})

X_train_res, y_train_res = smote.fit_resample(X_train_trf, y_train)
```

```python
pd.Series(y_train_res).value_counts()
```

##### undersampling of class "medium"

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(sampling_strategy={"medium": 600})

X_train_res, y_train_res = rus.fit_resample(X_train_res, y_train_res)
```

```python
pd.Series(y_train_res).value_counts()
```

```python
from sklearn.utils.class_weight import compute_class_weight

c_weights = compute_class_weight("balanced", classes=np.unique(y_train_res), y=y_train_res)

class_weights = dict(zip(["high", "low", "medium"], c_weights))
```

```python
class_weights
```

### DecisionTree

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(class_weight=class_weights)
clf.fit(X_train_res, y_train_res)
```

```python
from sklearn.metrics import accuracy_score, f1_score

y_pred = clf.predict(X_test_trf)

acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average="micro")
```

```python
print(f"accuracy: {acc}")
print(f"f1 score: {f1}")
```

```python
from sklearn.model_selection import GridSearchCV

clf = DecisionTreeClassifier(class_weight=class_weights)
params = {"criterion": ["gini", "entropy"], "max_depth": [15, 20, None], "min_samples_split": [2, 5]}

grid = GridSearchCV(estimator=clf, param_grid=params, cv=5, n_jobs=4)
grid.fit(X_train_res, y_train_res)
```

```python
best_estimator = grid.best_estimator_
best_params = grid.best_params_
best_score = grid.best_score_


print(best_params)
print(best_score)
```

```python
clf = DecisionTreeClassifier(**best_params, class_weight=class_weights)
clf.fit(X_train_res, y_train_res)


y_pred = clf.predict(X_test_trf)

acc = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average="micro")
```

```python
print(f"accuracy: {acc}")
print(f"f1 score: {f1}")
```

### RandomForest

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(class_weight=class_weights)

params = {"n_estimators": [400, 500, 600], "criterion": ["gini"], "max_depth": [14, 15, 20]}

grid = GridSearchCV(estimator=clf, param_grid=params, cv=5, n_jobs=4)
grid.fit(X_train_res, y_train_res)
```

```python
best_estimator = grid.best_estimator_
best_params = grid.best_params_
best_score = grid.best_score_


print(best_params)
print(best_score)
```

```python
clf = RandomForestClassifier(**best_params, class_weight=class_weights)

clf.fit(X_train_res, y_train_res)
y_pred = clf.predict(X_test_trf)
```

```python
print(y_pred[:10])
print(y_test[:10])
```

```python
from sklearn.metrics import classification_report


cr = classification_report(y_test, y_pred)
print(cr)
```
